    <br><b>check_win(self)</b>, ki pregleda ali je kdo zmagal in napiše točno kdo je zmagal
    <br>
    <br><b>reset(self)</b>, ki pobriše že igrane X in O

    <br><b>print_current_position(self)</b>, ki izriše trenutno pozicijo igre v obliki kvadrata, kjer je vsaka celica označena s trenutnimi znaki (za izgled si poglej "Otput" primera spodaj).
