````python
import matplotlib.pyplot as plt
import math
from matplotlib.animation import FuncAnimation

fig = plt.figure(figsize=(6,6,))

ax = plt.subplot(1, 1, 1)

x = [math.radians(i) for i in range(-180, 180)]
x_korak = x[-1] - x[-2]
cos = [math.cos(i) for i in x]
sin = [math.sin(i) for i in x]

def update(frame):
    global x, x_korak, ax, c, s
    del x[0]
    x.append(x[-1] + x_korak)
    cos = [math.cos(i) for i in x]
    sin = [math.sin(i) for i in x]
    
    ax.clear()
    ax.plot(x, cos)
    ax.plot(x, sin)
    ax.grid()
    
animation = FuncAnimation(fig, update, interval=4)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib notebook

n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.axes([0.025,0.025,0.95,0.95])
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

for x,y in zip(X,Y2):
    plt.text(x+0.4, -y-0.05, '%.2f' % y, ha='center', va= 'top')

plt.xlim(-.5,n), plt.xticks([])
plt.ylim(-1.25,+1.25), plt.yticks([])

plt.show()
````

Vaša naloga je, da napravite sledeč graf:



Velikost grafa naj bo 8x8.

Podatke **sin1** ("normalna sinus krivulja") prikažite z navadno črto, ki naj bo črne barve. Njeno ime naj bo **sinus**.

Podatke **sin2** ("podatki z vključenimi motnjami") prikažite kot točkovne vrednosti. Naj bodo rdeče barve. Zmanjšajte njihovo transparentnost (alpha). Ime podatkov naj bo "merjeni podatki"

**x_ticks** naj bodo vsakih 50 enot na x osi.

**y_ticks** naj bodo pri minimalni vrednosti *sin2* podatkov, -1, -0.5, 0, 0.5, 1, maxinalni vrednosti *sin2* podatkov. Minimalna vrednost in maximalna vrednost naj bosta podani na 3 decimalna mesta natančno.

Prikažite legendo in pomožne črte.

Za pomoč naj vam bo matplotlib dokumentacija.

# http://collabedit.com/br5qt

````python
import random
import math
import matplotlib.pyplot as plt

x = [(deg) for deg in range(0, 2*360)]
sin1 = [math.sin(math.radians(i)) for i in x]
sin2 = [i+(random.randint(-20,20)/100) for i in sin1]

fig = plt.figure(figsize=(8,8))

plt.plot(x, sin1, color="black", label="sinus")
plt.scatter(x, sin2, color="red", label="merjeni podatki", alpha=0.1)

x_ticks = x[::50]
plt.xticks(x_ticks)

y_ticks = [min(sin2), -1, -0.5, 0, 0.5, 1, max(sin2)]
y_ticks_notations = [f"{min(sin2):.3f}", "-1", "-0.5", "0", "0.5", "1", f"{max(sin2):.3f}"]
plt.yticks(y_ticks, y_ticks_notations)

plt.grid()

plt.legend()

plt.show()
````
